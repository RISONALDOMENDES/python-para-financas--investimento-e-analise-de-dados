### Qual é a distribuição normal? 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
normal_dis = np.random.normal(10,5,40)
normal_dis
```

desvio padrão (unidimensional)

```python
std_norm = np.random.standard_normal(40)

std_norm
```

desvio padrão (bi-dimensional)

```python
# dobra-se o numero de parenteses
std_norm_2 = np.random.standard_normal((4, 10))

std_norm_2
```

mostrando o desvio padrão em um grafico

```python
plt.plot(std_norm)
plt.show()
```
